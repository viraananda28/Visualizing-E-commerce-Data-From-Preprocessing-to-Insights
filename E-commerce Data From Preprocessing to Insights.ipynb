## Columns Renaming

```python
rename_mapper = {col:col.replace(" ", "_").lower() for col in df.columns}
df.rename(
    columns=rename_mapper,
    inplace=True
)

df.head()
```

## To Datetime datatype

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df[["order_date", "ship_date"]].head().info()
```

```python
df = df.sort_values(
    by=["order_id", "order_date", "ship_date"],
    ascending=[True, True, True],
    ignore_index=True
)

df.head(20)
```

## Remove "$" in ["sales", "profit", "shipping_cost" ]

```python
dollar_cols = ["sales", "profit", "shipping_cost"]
for col in dollar_cols:
  df[col] = (
      df[col]
      .str.replace("[^\d.]","", regex =True)
  )
  df.rename(columns={col: col+"($)"}, inplace=True)
print("Formatting & Renaming, Done!")
```

## Monthly Sales Trends

```python
monthly_sum_sales = df.groupby("months").agg({"sales($)": "sum"})
monthly_sum_sales
```

```python
#Memberikan urutan utk bulan
ordered_months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

Month_CI = pd.CategoricalIndex(
    monthly_sum_sales.index,
    ordered=True,
    categories=ordered_months
)

Month_CI
```

```python
monthly_sum_sales.index = Month_CI
monthly_sum_sales = monthly_sum_sales.sort_index()
monthly_sum_sales
```

```python
import matplotlib.ticker as ticker
```

```python
# Create line chart for monthly sales
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(
    monthly_sum_sales.index,
    monthly_sum_sales["sales($)"],
    marker='*',
    mfc='red', # marker face color
    mec='red', # marker edge color
    linestyle='-'
)

ax.set_title('Monthly Sales Trends')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.grid(True)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '$' + '{:,.2f}'.format(x/1000) + 'K'))

# Show the line chart
fig.show()
```

## 1.&nbsp; **Order Date vs. Sales/Profit**

```python
# Convert 'order_date' to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Set the figure size
plt.figure(figsize=(14, 7))

# Plot Sales over Time
sns.lineplot(data=df, x='order_date', y='sales($)', label='Sales')
sns.lineplot(data=df, x='order_date', y='profit($)', label='Profit')

plt.title('Sales and Profit Over Time')
plt.xlabel('Order Date')
plt.ylabel('Amount ($)')
plt.legend()
plt.show()
```

Insight: Penjualan berada di tingkat yang lebih tinggi dari pada keuntungan. Penjualan dan keuntungan menunjukkan tren musiman dengan fluktuasi yang cukup signifikan dari waktu ke waktu.

## 2.&nbsp; **Product Category vs. Sales/Profit**

```python
plt.figure(figsize=(10, 6))

# Plot Sales by Product Category
sns.barplot(data=df, x='product_category', y='sales($)', ci=None)

plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))

# Plot Profit by Product Category
sns.barplot(data=df, x='product_category', y='profit($)', ci=None)

plt.title('Profit by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)
plt.show()
```

Insight: Kategori produk tertentu, seperti "TFashion," cenderung menghasilkan penjualan dan keuntungan yang lebih tinggi dibandingkan kategori lainnya.
Strategi pemasaran dan penjualan bisa difokuskan pada kategori yang memberikan keuntungan lebih tinggi untuk memaksimalkan keuntungan.

# 3**. Sales vs. Quantity/Discount**

```python
plt.figure(figsize=(10, 6))

# Scatter plot of Sales vs. Quantity
sns.scatterplot(data=df, x='quantity', y='sales($)', hue='product_category', alpha=0.7)

plt.title('Sales vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Sales ($)')
plt.legend(title='Product Category')
plt.show()

plt.figure(figsize=(10, 6))

# Scatter plot of Sales vs. Discount
sns.scatterplot(data=df, x='discount', y='sales($)', hue='product_category', alpha=0.7)

plt.title('Sales vs. Discount')
plt.xlabel('Discount')
plt.ylabel('Sales ($)')
plt.legend(title='Product Category')
plt.show()
```

Insight: Kategori produk untuk Fashion dan elektrik selalu berada di tingkat penjualan dan keuntungan yang paling tinggi dibanding yang lainnya.

## 4.&nbsp; **Profit vs. Shipping Cost**

```python
plt.figure(figsize=(10, 6))

# Scatter plot of Profit vs. Shipping Cost
sns.scatterplot(data=df, x='shipping_cost($)', y='profit($)', hue='region', alpha=0.7)

plt.title('Profit vs. Shipping Cost')
plt.xlabel('Shipping Cost ($)')
plt.ylabel('Profit ($)')
plt.legend(title='Region')
plt.show()
```

## 7.&nbsp; **City, State, and Country-wise Analysis**

```python
plt.figure(figsize=(14, 7))

# Bar chart of Sales by City
city_sales = df.groupby('city')['sales($)'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=city_sales.index, y=city_sales.values, palette='viridis')

plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.show()
```

Insight: Beberapa kota seperti New York, Los Angeles, Philadelphia, dan San Fransisco menghasilkan penjualan yang jauh lebih tinggi dibandingkan kota lain, menunjukkan adanya konsentrasi pasar di lokasi tertentu.
Perusahaan dapat fokus pada kota dengan penjualan tertinggi agar menjadi strategi yang efektif untuk meningkatkan penjualan secara keseluruhan.

## 8.&nbsp; **Customer-wise Analysis**

```python
plt.figure(figsize=(14, 7))

# Bar chart of Top 10 Customers by Sales
top_customers = df.groupby('customer_name')['sales($)'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_customers.index, y=top_customers.values, palette='coolwarm')

plt.title('Top 10 Customers by Sales')
plt.xlabel('Customer Name')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.show()
```
